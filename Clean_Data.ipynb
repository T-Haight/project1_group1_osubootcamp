```python
# Import Libraries and Dependencies
import pandas as pd
import requests
```

```python
# Download the December 2021-2023 data downloaded from https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page
years = range(2021, 2024)
months = range(12, 13)
for year in years:
    for month in months:
        url = f'https://d37ci6vzurychx.cloudfront.net/trip-data/fhvhv_tripdata_{year}-{month:02}.parquet'
        response = requests.get(url)
        if response.status_code == 200:
            with open(f'fhvhv_tripdata_{year}-{month:02}.parquet', 'wb') as f:
                f.write(response.content)
        else:
            print(f'Failed to download {url}')
```

```python
# Read the fhvhv_tripdata parquet files in dataframes 
Dec_21_df = pd.read_parquet(f'fhvhv_tripdata_2021-12.parquet')
Dec_22_df = pd.read_parquet(f'fhvhv_tripdata_2022-12.parquet')
Dec_23_df = pd.read_parquet(f'fhvhv_tripdata_2023-12.parquet')

#  Display the first 5 rows of the Dec dataframes
display(Dec_21_df.head())
display(Dec_22_df.head())
display(Dec_23_df.head())
```

```python
# View column data types
Dec_23_df.info()
```

```python
# Extract important columns
cleaned_21_df = Dec_21_df[['hvfhs_license_num', 'pickup_datetime', 'dropoff_datetime', 'PULocationID', 'DOLocationID', 'trip_miles', 'trip_time', 'base_passenger_fare', 'tolls', 'bcf', 'sales_tax'
                , 'congestion_surcharge', 'airport_fee', 'tips', 'driver_pay']].copy()
cleaned_22_df = Dec_22_df[['hvfhs_license_num', 'pickup_datetime', 'dropoff_datetime', 'PULocationID', 'DOLocationID', 'trip_miles', 'trip_time', 'base_passenger_fare', 'tolls', 'bcf', 'sales_tax'
                , 'congestion_surcharge', 'airport_fee', 'tips', 'driver_pay']].copy()
cleaned_23_df = Dec_23_df[['hvfhs_license_num', 'pickup_datetime', 'dropoff_datetime', 'PULocationID', 'DOLocationID', 'trip_miles', 'trip_time', 'base_passenger_fare', 'tolls', 'bcf', 'sales_tax'
                , 'congestion_surcharge', 'airport_fee', 'tips', 'driver_pay']].copy()

# Display the first 5 rows of the cleaned_23_df dataframe
cleaned_23_df.head()
```

```python
# Calculate the total passenger cost
cleaned_21_df['Total_Passenger_Cost'] = cleaned_21_df['base_passenger_fare'] + cleaned_21_df['tolls'] + cleaned_21_df['bcf'] + cleaned_21_df['sales_tax'] + cleaned_21_df['congestion_surcharge'] + cleaned_21_df['airport_fee'] + cleaned_21_df['tips']
cleaned_22_df['Total_Passenger_Cost'] = cleaned_22_df['base_passenger_fare'] + cleaned_22_df['tolls'] + cleaned_22_df['bcf'] + cleaned_22_df['sales_tax'] + cleaned_22_df['congestion_surcharge'] + cleaned_22_df['airport_fee'] + cleaned_22_df['tips']
cleaned_23_df['Total_Passenger_Cost'] = cleaned_23_df['base_passenger_fare'] + cleaned_23_df['tolls'] + cleaned_23_df['bcf'] + cleaned_23_df['sales_tax'] + cleaned_23_df['congestion_surcharge'] + cleaned_23_df['airport_fee'] + cleaned_23_df['tips']

# Display the first 5 rows of the cleaned_23_df dataframe
cleaned_23_df.head()
```

```python
# Filter the data for Lyft
lyft_21_df = cleaned_21_df.loc[cleaned_21_df['hvfhs_license_num'] == 'HV0005'].copy()
lyft_22_df = cleaned_22_df.loc[cleaned_22_df['hvfhs_license_num'] == 'HV0005'].copy()
lyft_23_df = cleaned_23_df.loc[cleaned_23_df['hvfhs_license_num'] == 'HV0005'].copy()

# Display the first 5 rows of the lyft_23_df dataframe
lyft_23_df.head()
```

```python
# Remove the hvfhs_license_num column, since it is no longer needed
lyft_21_df.drop(columns='hvfhs_license_num', inplace=True)
lyft_22_df.drop(columns='hvfhs_license_num', inplace=True)
lyft_23_df.drop(columns='hvfhs_license_num', inplace=True)

# Display the first 5 rows of the lyft_23_df dataframe
lyft_23_df.head()
```

```python
# Sort values by pickup_datetime 
lyft_21_df.sort_values("pickup_datetime", inplace=True)
lyft_22_df.sort_values("pickup_datetime", inplace=True)
lyft_23_df.sort_values("pickup_datetime", inplace=True)

# Reset the index
lyft_21_df.reset_index(drop=True, inplace=True)
lyft_22_df.reset_index(drop=True, inplace=True)
lyft_23_df.reset_index(drop=True, inplace=True)

# Display the first 5 rows of the lyft_23_df dataframe
lyft_23_df.head()
```

```python
# Save the data to a parquet files
lyft_21_df.to_parquet(f'lyft_Dec_21.parquet')
lyft_22_df.to_parquet(f'lyft_Dec_22.parquet')
lyft_23_df.to_parquet(f'lyft_Dec_23.parquet')
```
