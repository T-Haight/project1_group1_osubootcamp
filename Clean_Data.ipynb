```python
import requests
import json
import pandas as pd

# Import dotenv package for setting environment variables 
from dotenv import load_dotenv

# Import os package
import os
```

```python
df = pd.read_parquet('../Downloads/fhvhv_tripdata_2023-12.parquet')
df.sort_index(inplace=True)
display(df.head())
df.tail()
```

```python
cleaned_df = df[['hvfhs_license_num', 'pickup_datetime', 'dropoff_datetime', 'PULocationID', 'DOLocationID', 'trip_miles', 'trip_time', 'base_passenger_fare', 'tolls', 'bcf', 'sales_tax'
                , 'congestion_surcharge', 'airport_fee', 'tips', 'driver_pay']]
cleaned_df.head()
```

```python
cleaned_df['Total_Passenger_Cost'] = cleaned_df['base_passenger_fare'] + cleaned_df['tolls'] + cleaned_df['bcf'] + cleaned_df['sales_tax'] + cleaned_df['congestion_surcharge'] + cleaned_df['airport_fee'] + cleaned_df['tips']
cleaned_df.head()
```

```python
cleaned_df.info()
```

```python
grouped_df = cleaned_df.groupby('hvfhs_license_num').agg({'trip_miles': 'sum', 'trip_time': 'sum', 'Total_Passenger_Cost': 'sum', 'driver_pay': 'sum'})
grouped_df.head()
```
